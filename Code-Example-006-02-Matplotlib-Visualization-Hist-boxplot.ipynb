```python
import numpy as np
import matplotlib.pyplot as plt
```

A 2006 paper (Low Birth Weight, a Risk Factor for Cardiovascular Diseases in Later Life, Is Already Associated with Elevated Fetal Glycosylated Hemoglobin at Birth) showed a histogram of the birth weights of newborns in their sample.

Scientists collected the weight of 10000 new born babies. 


```python

# Let us here assume that the data has a mean of 3500 is with a standard deviation of 
mu, sigma = 3500, 500 # mean and standard deviation
weights_of_new_born_babies_in_gr = np.round(np.random.normal(mu, sigma, 10000))

# Print 10 example of the weights. 
print("10 example weights: ", weights_of_new_born_babies_in_gr[:10])


# the histogram of the data
plt.hist(weights_of_new_born_babies_in_gr, 20, density=False, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.grid(True)
plt.show()
```

We can see that the shape of this distribution has a symetic shape. 

# Histogram

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```

# Boxplot

Boxplot shows where the max, min, median, upper and lower quartiles are. 
We can also see on Boxplots the outliers. Most libraries provide automated outlier detection on boxplot.

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Boxplot of New born babies')
ax1.boxplot(weights_of_new_born_babies_in_gr)
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()
ax3.set_title('Changed Outlier Symbols')
ax3.boxplot(weights_of_new_born_babies_in_gr, flierprops=green_diamond)
```

```python
fig4, ax4 = plt.subplots()
ax4.set_title('Hide Outlier Points')
ax4.boxplot(weights_of_new_born_babies_in_gr, showfliers=False)
```

```python
# Horizontal Boxplot
red_square = dict(markerfacecolor='r', marker='s')
fig5, ax5 = plt.subplots()
ax5.set_title('Horizontal Boxes')
ax5.boxplot(weights_of_new_born_babies_in_gr, vert=False, flierprops=red_square)
```
